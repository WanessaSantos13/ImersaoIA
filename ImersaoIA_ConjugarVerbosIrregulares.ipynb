```python
def embed_fn(Infinitivo, PassadoSimples, ParticipioPassado, Significado):
    return genai.embed_content(model=model,
                              content=[PassadoSimples, ParticipioPassado, Significado],
                               title=Infinitivo,
                               task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
# prompt: Usando o DataFrame df: gerar_consulta

def gerar_consulta(df, termo):
    # Filtra o dataframe para obter somente as linhas que contenham o termo na coluna "Infinitivo".
    df_filtrado = df[df["Infinitivo"].str.contains(termo, case=False)]

    # Retorna o dataframe filtrado.
    return df_filtrado


```
